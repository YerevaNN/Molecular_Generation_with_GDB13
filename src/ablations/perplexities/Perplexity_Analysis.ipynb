```python
import random
import json
import pandas as pd
import selfies as sf
from tqdm import tqdm
from rdkit import Chem
import matplotlib.pyplot as plt
```

```python
def make_canonical_smiles(selfies):
    canon_smiles = None
    smiles = sf.decoder(selfies)

    try:
        canon_smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    except:
        pass  

    return canon_smiles
```

```python
make_canonical_smiles("[C][O][C][=Branch1][C][=O][C][C][Branch1][C][C][N][C][C][C][=Branch1][C][=O][C][Ring1][=Branch1]")
```

## Perplexities of Sas Selfies

```python
gen_sas_df_aa = pd.read_csv("data-bin-gen-sas-selfies-aa-10K_ppl.csv", names=["ppl", "selfies"])
gen_sas_df_aa = gen_sas_df_aa.dropna()
gen_sas_df_aa["smiles"] = gen_sas_df_aa["selfies"].apply(make_canonical_smiles)
print("Lenght", len(gen_sas_df_aa))
gen_sas_df_aa.head()
```
